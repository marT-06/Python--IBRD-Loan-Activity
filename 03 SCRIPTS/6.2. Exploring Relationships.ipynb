# Exploring Relationships

## Correlations

```python
# Create a correlation matrix using pandas

df.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# adding figure size, labels,and a legend

f = plt.figure(figsize=(19, 19))
plt.matshow(sub.corr(), fignum=f.number)
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=14, rotation=70) 
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) 
cb = plt.colorbar() 
cb.ax.tick_params(labelsize=14) 
plt.title('Correlation Matrix', fontsize=14) 
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(19,19))

# Create the correlation heatmap in seaborn
# by applying a heatmap onto the correlation matrix and the subplots defined above

corr = sns.heatmap(sub.corr(), annot = True, ax = ax)

# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

- A -0.35 coefficient suggests negative correlation between Interest Rates and the Loans Held, which is moderate-to-weak.


- A higher Original Principal Amount will result in a bigger Disbursed Amount, according to the strong positive coefficient (0.93). That makes sense given that the IBRD is distributing funds to assist nations with their projects.


- There is a very strong positive correlation between the Borrower's Obligation and Due to IBRD (0.99), as well as between the Disbursed Amount and Due to IBR (0.75), demonstrating that the more money IBRD disburses to the Borrower, the more obligations he will have and the greater responsibility he will have to IBRD.

## Scatterplots

```python
 
sns.lmplot(x = 'Disbursed Amount', y = 'Exchange Adjustment', data = df)
```

- This scatterplot shows a weak negative relationship. There are many points that are far from the trend line and the trend line is not particularly steep. 
- These two columns' relationship displays two distinct upwards and downward trends, with the downward trend being stronger. This is a consequence of exchange rate variations that have caused both the disbursed amounts to rise or fall.

```python
# Create a scatterplot in seaborn

sns.lmplot(x = 'Disbursed Amount', y = 'Original Principal Amount', data = df)
```

- We can observe from the scatterplot that there is a strong positive correlation because the trend line is steep and most of the data points clustered around the line.
- We can also spot some outliers on the scatterplot.
- As Original Principal Amount increases so does the Disbured Amount as well.

```python
sns.lmplot(x = 'Borrowers Obligation', y = 'Due to IBRD', data = df)
```

- The scatterplot shows a strong positive relationship between Borrower's obligation and the outstanding amount. When one increases so does the other as well.
- Most data points are close to the trend, so we can easily see the outlier in the right-up corner.

```python
sns.lmplot(x = 'Disbursed Amount', y = 'Due to IBRD', data = df)
```

- Despite being positive, the connection between the "Disbursed Amount" and the "Due to IBRD" is not entirely linear. 
- Although the great majority of the data points show an upward trend, we can also see a few data points at the "Due to IBRD" = O coordinate that are not following an upward trend. 
- Those points can refer to contracts that have already been closed and disbursed payments that are no longer due to IBRD.

# Categorical Plot

```python
# Creating a random sample of sub dataframe to use for scatterplot
np.random.seed(4)
dev = np.random.rand(len(sub)) <= 0.90
big = sub[dev]
small = sub[~dev]
```

```python
sub.shape
```

```python
len(small)+len(big)
```

```python
len(small)
```

```python
# Use a histogram to visualize the distribution of the variables.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(small['Interest Rate'], bins = 50, kde = True)
```

```python

small.loc[(small['Interest Rate'] == 0), 'Interest Rate Category'] = 'Zero Interest Rate'

small.loc[(small['Interest Rate'] > 0) & (small['Interest Rate'] <= 3), 'Interest Rate Category'] = 'Low Interest Rate'
       
small.loc[(small['Interest Rate'] > 3) & (small['Interest Rate'] < 9), 'Interest Rate Category'] = 'Middle Interest Rate'

small.loc[small['Interest Rate'] >= 9, 'Interest Rate Category'] = 'High Interest Rate'

```

```python
small['Interest Rate Category'].value_counts(dropna = False)
```

```python
 %%time

sns.set(style="ticks")

g = sns.catplot(x='Original Principal Amount', y='Interest Rate', hue='Interest Rate Category', data=small)
```

- We can observe from this catplot that all of the Interest Categories are present through all the Original Principal Amounts. 
- It is interesting that for the High Original Principal Amount we have a higher density of Low Interest rates, we can see more red dots on the lower-right part of the visual.

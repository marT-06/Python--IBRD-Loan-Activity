# Drop missing values from date columns

```python
#droping missing values from date columns

df=df.dropna(subset=['Agreement Signing Date','Effective Date (Most Recent)','Closed Date (Most Recent)'])
```

# Extracting Years

```python
# extracting years from date columns

df['Agreement Year'] = df['Agreement Signing Date'].astype(str).str[:4]

df['Effective Year'] = df['Effective Date (Most Recent)'].astype(str).str[:4]

df['Closed Year'] = df['Closed Date (Most Recent)'].astype(str).str[:4]

df
```

```python
# New df including years as of 2000 onwards

df_new = df[(df['Agreement Year'] > '1999') & (df['Effective Year'] > '1999') &(df['Closed Year'] > '1999')]
df_new
```

```python
#dropping years column

df_clean = df_new.drop(columns = [ 'Agreement Year','Effective Year', 'Closed Year'])
df_clean
```

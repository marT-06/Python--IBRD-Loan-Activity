# Normalising data

```python
# define min max scaler

scaler = MinMaxScaler()
```

```python
# transform data

df_num[['Interest Rate','Original Principal Amount', 
                'Cancelled Amount', 'Undisbursed Amount','Disbursed Amount', 
                'Repaid to IBRD', 'Due to IBRD', 'Exchange Adjustment', 
                'Borrowers Obligation', 'Loans Held']]= scaler.fit_transform(df_num[['Interest Rate', 'Original Principal Amount',
                                                                                              'Cancelled Amount', 'Undisbursed Amount',
                                                                                              'Disbursed Amount', 'Repaid to IBRD', 'Due to IBRD',
                                                                                              'Exchange Adjustment', 'Borrowers Obligation', 'Loans Held']])
```

# The elbow technique

```python
# Defining the range of potential clusters in the data

num_cl = range(1, 10) 

# Defining k-means clusters in the range assigned above.

kmeans = [KMeans(n_clusters=i) for i in num_cl] 
```

```python
# Creating a score that represents a rate of variation for the given cluster option

score = [kmeans[i].fit(df_num).score(df_num) for i in range(len(kmeans))] 

score
```

```python
# Plotting the elbow curve using PyLab.

pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

- The curve is showing a jump from two to four on the x-axis, after which it straightens out. 
- This means that the optimal count for clusters is 2.

# k-means clustering

```python
# Create the k-means object.

kmeans = KMeans(n_clusters = 2) 
```

```python
# Fit the k-means object to the data.

kmeans.fit(df_num)
```

```python
df_num['clusters'] = kmeans.fit_predict(df_num)
```

```python
df_num.head()
```

```python
df_num['clusters'].value_counts()
```

```python
# Plot the clusters for the "Borrowers Obligation" and "Due to IBRD" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['Due to IBRD'], y=df['Borrowers Obligation'], hue=kmeans.labels_, s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm
# `s` represents the size of the points you want to see in the plot

ax.grid(False) # This removes the grid from the background
plt.xlabel('Due to IBRD') # Label x-axis
plt.ylabel('Borrowers Obligation') # Label y-axis
plt.show()
```

- The first cluster, in dark pink (coded as “3” in the legend), has only 2 visible points which are representing really low amount rates for both, Due to IBRD and Borrowers Obligation.

- The second cluster, in purple (coded as “2” in the legend), includes points with relatively small amount rates of Due to IBRD and relatively small amount rates of Borrowers Obligations.

- The third cluster, in light pink (coded as “0” in the legend), is the most populated cluster. It gathers the data points with the highest amount rates of Due to IBRD and highest rates of Borrowers Obligations. 

- The cluster in pink (coded as “1” in the legend) is not visible at all.

```python
# Plot the clusters for the "Original Principal Amount" and "Due to IBRD" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['Due to IBRD'], y=df['Original Principal Amount'], hue=kmeans.labels_, s=100) 

ax.grid(False) 
plt.xlabel('Due to IBRD')
plt.ylabel('Original Principal Amount')
plt.show()
```

- The first cluster, in dark pink (coded as “3” in the legend), has only a few visible points which are representing the lowest amount rates for both, Due to IBRD and Original Principal Amount.

- The second in pink (coded as “1” in the legend) has a few visible data points which are relatively low for both variables.

- The third cluster, in purple (coded as “2” in the legend)is a very populated cluster. It includes points with relatively small amount rates of Due to IBRD, some of which are zero, and relatively small amount rates of Borrowers Obligations.

- The fourth cluster, in light pink (coded as “0” in the legend), is showing a positive trend. It gathers the data points with the highest amount rates of Due to IBRD and the highest Original Principal Amount rates. 

```python
df_num.loc[df_num['clusters'] == 3, 'cluster'] = 'dark purple'
df_num.loc[df_num['clusters'] == 2, 'cluster'] = 'purple'
df_num.loc[df_num['clusters'] == 1, 'cluster'] = 'pink'
df_num.loc[df_num['clusters'] == 0, 'cluster'] = 'light pink'
```

```python
df_num.groupby('cluster').agg({'Original Principal Amount':['mean', 'median'],
                             'Due to IBRD':['mean', 'median'],
                             'Borrowers Obligation':['mean', 'median']})
```

- The light pink cluster has significantly higher means in all categories. This once again confirms that this cluster contains the data points with the highest rates of amounts for every variable.


- The purple cluster has similar rates of amounts as pink and dark purple clusters, for all the categories.

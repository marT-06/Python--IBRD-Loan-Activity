## Deleting columns that won't be used in Analysis 

- the columns that are irrelevant for the analysis will be dropped

```python
# dropping columns we don't need for the analysis

df_clean = df_orig.drop(columns = [ 'End of Period','Country Code','Guarantor Country Code',
                                   'Currency of Commitment', 'Project ID', 'Project Name','Sold 3rd Party', 
                                   'Repaid 3rd Party', 'Due 3rd Party', 'First Repayment Date',
                                   'Last Repayment Date', 'Board Approval Date','Last Disbursement Date'])
```

- Droppedd End of Period because it represents the date when the snapshot was taken (31.07.2022. for all records)
- Other columns are dropped because they are not relevant for the analysis

## Missing values

```python
# counting null values

null_values = pd.DataFrame({'columns':pd.isnull(df_clean).sum().index, 
                            'count':pd.isnull(df_clean).sum().values})

null_values.loc[null_values['count']>0,].sort_values(by='count', ascending=False)
```

```python
# checking for the percentage of NaNs

for col in (df_clean.columns):
    pct_missing = np.mean(df_clean[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100,2)))
```

### Deleting missing values

```python
#checking the shape

df_clean.shape
```

```python
# deleting NaNs from columns

df_clean = df_clean.dropna(subset=[ 'Borrower', 'Guarantor', 'Interest Rate','Agreement Signing Date',
                                   'Effective Date (Most Recent)','Closed Date (Most Recent)'])
```

```python
#checking the shape

df_clean.shape
```

## Duplicates

```python
# Look for full duplicates 
# No duplcates found

df_dups = df_clean[df_clean.duplicated()]
df_dups
```

## Changing data types

```python
# renaming "Borrower's Obligation" into "Borrowers Obligation" because of ' sign

df_clean.rename(columns = {"Borrower's Obligation" : 'Borrowers Obligation'}, inplace = True)
```

```python
# change all text columns into a string 

text = 'Loan Number','Region','Country', 'Borrower','Guarantor','Loan Type','Loan Status'

for x in text:
    df_clean[x] = df_clean[x].astype('string')
```

```python
# changing all columns representing dates to a datetime type

dates = ['Agreement Signing Date', 'Effective Date (Most Recent)', 'Closed Date (Most Recent)']

for x in dates:
    df_clean[x] = pd.to_datetime(df_clean[x])
```

## Mixed types

```python
# check for mixed types

for col in df_clean.columns.tolist():
  weird = (df_clean[[col]].applymap(type) != df_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_clean[weird]) > 0:
    print (col)
```
